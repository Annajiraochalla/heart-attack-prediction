```python
#importing neccessary libraries and modules
import warnings
import pandas as pd
from sklearn.svm import SVC
from matplotlib import pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.feature_selection import mutual_info_classif as MIC
from sklearn.metrics import confusion_matrix, f1_score, log_loss, accuracy_score
```

```python
#reading the dataset
dataset = pd.read_csv('heart.csv')
dataset
```

```python
dataset.describe()
```

```python
#separting the input and output
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X
```

```python
y
```

```python
#spliting the original dataset into train and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

#using standard scaler function
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

#using four different models
classifier_1 = DecisionTreeClassifier(criterion='entropy')
classifier_2 = LogisticRegression()
classifier_3 = SVC(kernel = 'linear', random_state = 0)
classifier_4 = GaussianNB()

#training the models
classifier_1.fit(X_train, y_train)
classifier_2.fit(X_train, y_train)
classifier_3.fit(X_train, y_train)
classifier_4.fit(X_train, y_train);
```

```python
#evaluting metrics of each model
y_pred_1 = classifier_1.predict(X_test)
y_pred_2 = classifier_2.predict(X_test)
y_pred_3 = classifier_3.predict(X_test)
y_pred_4 = classifier_4.predict(X_test)

#confusion matrix
cm_1 = confusion_matrix(y_test, y_pred_1)
cm_2 = confusion_matrix(y_test, y_pred_2)
cm_3 = confusion_matrix(y_test, y_pred_3)
cm_4 = confusion_matrix(y_test, y_pred_4)

#f1 score
f1_1 = f1_score(y_test,y_pred_1)
f1_2 = f1_score(y_test,y_pred_2)
f1_3 = f1_score(y_test,y_pred_3)
f1_4 = f1_score(y_test,y_pred_4)

#log loss
ll_1 = log_loss(y_test, y_pred_1, eps=1e-15, normalize=True)
ll_2 = log_loss(y_test, y_pred_2, eps=1e-15, normalize=True)
ll_3 = log_loss(y_test, y_pred_3, eps=1e-15, normalize=True)
ll_4 = log_loss(y_test, y_pred_4, eps=1e-15, normalize=True)

#accuracy score
acs_1 = accuracy_score(y_test, y_pred_1)
acs_2 = accuracy_score(y_test, y_pred_2)
acs_3 = accuracy_score(y_test, y_pred_3)
acs_4 = accuracy_score(y_test, y_pred_4)

print('For Original Dataset: \n')

print('For Decision Tree Classifier:')
print('Confusion Matrix:\n',cm_1)
print('F1 Score:', round(f1_1 * 100, 2))
print('Log Loss:', ll_1)
print ('Accuracy Score:', round(acs_1 * 100,2),'%')

print('\nFor Logistic Regression:')
print('Confusion Matrix:\n',cm_2)
print('F1 Score:', round(f1_2 * 100, 2))
print('Log Loss:', ll_2)
print ('Accuracy Score:', round(acs_2 * 100,2),'%')

print('\nFor Support Vector Machine Classifier:')
print('Confusion Matrix:\n',cm_3)
print('F1 Score:', round(f1_3 * 100, 2))
print('Log Loss:', ll_3)
print ('Accuracy Score:', round(acs_3 * 100,2),'%')

print('\nFor Gaussain Naive Bayes Classifier:')
print('Confusion Matrix:\n',cm_4)
print('F1 Score:', round(f1_4 * 100, 2))
print('Log Loss:', ll_4)
print ('Accuracy Score:', round(acs_4 * 100,2),'%')
```

```python
#k cross validation for standard accuracy
accuracy_1 = cross_val_score(estimator = classifier_1, X = X_train, y = y_train, cv = 10)
accuracy_2 = cross_val_score(estimator = classifier_2, X = X_train, y = y_train, cv = 10)
accuracy_3 = cross_val_score(estimator = classifier_3, X = X_train, y = y_train, cv = 10)
accuracy_4 = cross_val_score(estimator = classifier_4, X = X_train, y = y_train, cv = 10)

print('For Decision Tree Classifier:')
print("Mean Accuracy: {:.2f} %".format(accuracy_1.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy_1.std()*100))
print()
print('For Logistic Regression:')
print("Mean Accuracy: {:.2f} %".format(accuracy_2.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy_2.std()*100))
print()
print('For Support Vector Machine Classifier:')
print("Mean Accuracy: {:.2f} %".format(accuracy_3.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy_3.std()*100))
print()
print("For Gaussain Naive Bayes Classifier:")
print("Mean Accuracy: {:.2f} %".format(accuracy_4.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracy_4.std()*100))
```

```python
#hyperparameter turning

#to avoid warnings
warnings.filterwarnings('ignore')

#parameters for decision tree classifier
parameters_1 = [{'criterion': ['entropy','log_loss'], 'splitter': ['best', 'random']}]
#parameters for logistic regression
parameters_2 = [{'penalty': ['l1','l2','elasticnet'], 'solver' : ['lbfgs','liblinear','saga']}]
#parameter for svm classifier
parameters_3 = [{'C': [0.25, 0.5], 'kernel': ['rbf'], 'gamma': [0.1, 0.2, 0.3]}]
#parameter for gaussain naive bayes classifier
parameters_4 = [{'var_smoothing': [1e-9, 1e-6, 1e-3]}]

#performing grid search

#grid_search for decision tree classifier
grid_search_1 = GridSearchCV(estimator = classifier_1,
                           param_grid = parameters_1,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)

#grid_search for logistic regression
grid_search_2 = GridSearchCV(estimator = classifier_2,
                           param_grid = parameters_2,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)

#grid_search for svm classifier
grid_search_3 = GridSearchCV(estimator = classifier_3,
                           param_grid = parameters_3,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)

#grid_search for gaussain naive bayes classifier
grid_search_4 = GridSearchCV(estimator = classifier_4,
                           param_grid = parameters_4,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1)


grid_search_1.fit(X_train, y_train)
best_accuracy_1 = grid_search_1.best_score_
best_parameters_1 = grid_search_1.best_params_
print('For Decision Tree Classifier:')
print("Best Accuracy: {:.2f} %".format(best_accuracy_1*100))
print("Best Parameters:", best_parameters_1)
print()
grid_search_2.fit(X_train, y_train)
best_accuracy_2 = grid_search_2.best_score_
best_parameters_2 = grid_search_2.best_params_
print('For Logistic Regression:')
print("Best Accuracy: {:.2f} %".format(best_accuracy_2*100))
print("Best Parameters:", best_parameters_2)
print()
grid_search_3.fit(X_train, y_train)
best_accuracy_3 = grid_search_3.best_score_
best_parameters_3 = grid_search_3.best_params_
print('For Support Vector Machine Classifier:')
print("Best Accuracy: {:.2f} %".format(best_accuracy_3*100))
print("Best Parameters:", best_parameters_3)
print()
grid_search_4.fit(X_train, y_train)
best_accuracy_4 = grid_search_4.best_score_
best_parameters_4 = grid_search_4.best_params_
print("For Gaussain Naive Bayes Classifier:")
print("Best Accuracy: {:.2f} %".format(best_accuracy_4*100))
print("Best Parameters:", best_parameters_4)
```
